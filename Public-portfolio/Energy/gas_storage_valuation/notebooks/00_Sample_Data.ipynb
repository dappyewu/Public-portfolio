##  Data for testing from Yfinance
Not required if you have a sample spot history 

```python
import yfinance as yf
import pandas as pd

ticker = "TTF=F"

# Download data
df = yf.download(
    ticker,
    start="2021-01-01",
    end="2025-12-31",
    interval="1d"
)

df.head()
```

```python
spot = df.reset_index()[["Date", "Close"]]
spot.columns = ["date", "spot_eur_mwh"]

spot.to_csv("sample_spot_history_ttf_real.csv", index=False)
```
